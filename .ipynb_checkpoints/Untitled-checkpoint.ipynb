```python
import pandas as pd
```

```python
train_data = pd.read_csv('./data/train.csv',header=None)
result = pd.read_csv('./data/trainLabels.csv',header=None)
test_data = pd.read_csv('./data/test.csv',header=None)
train_data['result'] = result[0]

train_data.head()
```

```python
correlations = abs(train_data.corr()['result'].sort_values())
print('Most Positive Correlations:\n', abs(correlations).sort_values().head())
high_list = list(train_data.columns[abs(correlations >= 0.05)])
high_list.remove('result')
```

```python
print('=====SVM=====')
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

train_x = train_data[high_list]
x_train, x_test, y_train, y_test = train_test_split(train_x, result, test_size=0.1, random_state=600)
clf=svm.SVC(kernel='linear',C=0.1,gamma='auto')
clf.fit(x_train,y_train)
y_test_predict = clf.predict(x_test)
Ypred = clf.predict(x_test)
print("SVM_accuracy_score",accuracy_score(y_test, Ypred),'\n','\n')
```

```python
df = pd.read_csv('./data/test.csv',header=None)
train_x = df[high_list]
df2 = pd.DataFrame(clf.predict(train_x),columns = ['Solution'])
df2.index = range(1,len(df2)+1)
df2['Id'] = df2.index
print(df2)
df2.to_csv('./prediction.csv',index=False)
```
